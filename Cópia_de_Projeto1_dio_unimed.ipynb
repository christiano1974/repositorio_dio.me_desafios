```python
### importação dos pacotes
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("ggplot")
```

```python
### importação dos arquivos em excel
df1 = pd.read_excel('/content/Aracaju.xlsx')
df2 = pd.read_excel('/content/Fortaleza.xlsx')
df3 = pd.read_excel('/content/Natal.xlsx')
df4 = pd.read_excel('/content/Recife.xlsx')
df5 = pd.read_excel('/content/Salvador.xlsx')
```

```python
### ler as 5 primeira linhas do dataframe
df1.head()
```

```python
### concatenar as planilhas
df = pd.concat([df1,df2,df3,df4,df5])
df.sample(5)
```

```python
### verificar a dimensão do dataframe
df.shape
```

```python
### verificar se há dados faltantes Nan. Neste caso não tem dados faltantes.
df.isnull().sum()
```

```python
### verificar as variáveis e seus tipos.
df.dtypes
```

```python
### transformar a coluna LojaID de int64 (inteiro) para object (string)
df['LojaID'] = df['LojaID'].astype("object")
```

```python
### verificação se a mudança ocorreu com sucesso
df.dtypes
```

```python
### criar a coluna Receita.
df['Receita'] = df['Qtde'].mul(df['Vendas'])
```

```python
### verificar quais colunas
df.columns
```

```python
### ler as 5 primeiras linhas do dataframe
df.head()
```

```python
### Calcular a receita em relação as Cidades
df.groupby("Cidade")["Receita"].sum()
```

```python
### Calcular a receita em relação as LojaID
df.groupby("LojaID")["Receita"].sum()
```

```python
### Calcular a receita em relação ao ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
### criar as seguintes colunas ano_venda, mês_venda e dia_venda
df["Ano_Venda"] = df["Data"].dt.year
df["Mês_Venda"] = df["Data"].dt.month
df["Dia_Venda"] = df["Data"].dt.day
```

```python
df.head()
```

```python
df["Trimestre_Venda"] = df["Data"].dt.quarter
```

```python
df.head()
```

```python
### 3 maiores receitas.
df.nlargest(3,"Receita")
```

```python
### 3 piores receitas.
df.nsmallest(3,"Receita")
```

```python
df.dtypes
```

```python
df["Ano_Venda"] = df["Ano_Venda"].astype("object")
df["Mês_Venda"] = df["Mês_Venda"].astype("object")
df["Dia_Venda"] = df["Dia_Venda"].astype("object")
df["Trimestre_Venda"] = df["Trimestre_Venda"].astype("object")
```

```python
df.dtypes
```

```python
### Análise descretiva dos dados numéricos
df.describe()
```

```python
### Análise descretiva dos dado categórico - Cidade
### As maiores quantidades vendidas foram em Fortaleza e Salvador
df["Cidade"].unique()
freq_absoluta = [len(df[df["Cidade"]=='Aracaju']),len(df[df["Cidade"]=='Fortaleza']),len(df[df["Cidade"]=='Natal']),len(df[df["Cidade"]=='Recife']),len(df[df["Cidade"]=='Salvador'])]
freq_relativa = [round(100*k/894,2) for k in freq_absoluta]
pd.DataFrame({'freq_absoluta':freq_absoluta,'freq_relativa %':freq_relativa},index=['Aracaju','Fortaleza','Natal','Recife','Salvador'])
```

## Vistualização dos dados.

```python
### gráfico de colunas
plt.figure(figsize=(10,5))
df["LojaID"].value_counts(ascending = False).plot.bar(title = 'Receitas dos IDLojas.');
plt.xlabel('IDLojas')
plt.ylabel('Receita em reais')
plt.show()
```

```python
### gráfico de barras
plt.figure(figsize=(10,5))
df["LojaID"].value_counts(ascending = False).plot.barh(title = 'Receitas dos IDLojas.');
plt.ylabel('IDLojas')
plt.xlabel('Receita em reais')
plt.show()
```

```python
### gráfico de setores ou pizza
plt.figure(figsize = (10,7))
df.groupby(df["Data"].dt.year)["Receita"].sum().plot.pie(title = "Receitas de 2018 e 2019");
plt.legend();
```

```python
### gráfico de linha
df_2019 = df[df["Ano_Venda"] == 2019]
plt.figure(figsize=(10,5))
df_2019.groupby(df_2019["Mês_Venda"])["Qtde"].sum().plot(marker = "v")
plt.xlabel("Mês")
plt.ylabel("Receita em reais")
plt.legend()
plt.show()
```
